## Relative perm

```python
sw = np.linspace(0,1,100)
```

```python
sw
```

```python
krw = sw**3
```

```python
krw
```

```python
kro = (1-sw)**2
```

```python
kro
```

```python
#Step 1: Krw vs Sw plot
```

```python
plt.plot(sw,krw)
```

```python
# Step2: Kro vs Sw plot

plt.plot(sw,kro)
```

```python
#Step 3: plotting kro,krw vs sw on same plot
plt.plot(sw,krw)
plt.plot(sw,kro)
```

```python
# Step 4: Give x axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation,Sw(fraction)")
```

```python
# Step 5: Give y axis label

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")
```

```python
# Step 6: Grid

plt.plot(sw,krw)
plt.plot(sw,kro)

plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
```

```python
# Step 7: Legend

plt.plot(sw,krw,label="Krw(Water Relative Permeability)")
plt.plot(sw,kro,label="Kro(Oil Relative Permeability)")


plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 8: Change the size of the plot

plt.figure(figsize=(12,8))

plt.plot(sw,krw,label="Krw(Water Relative Permeability)")
plt.plot(sw,kro,label="Kro(Oil Relative Permeability)")


plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 9: Change the line color

plt.figure(figsize=(12,8))

plt.plot(sw,krw,label="Krw(Water Relative Permeability)",color='blue')
plt.plot(sw,kro,label="Kro(Oil Relative Permeability)",color="green")


plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 10: Change the line thickness

plt.figure(figsize=(12,8))

plt.plot(sw,krw,label="Krw(Water Relative Permeability)",color='blue',lw=4)
plt.plot(sw,kro,label="Kro(Oil Relative Permeability)",color="green",lw=0.7)


plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 11: Change the line type

plt.figure(figsize=(12,8))

plt.plot(sw,krw,"*",label="Krw(Water Relative Permeability)",color='blue')
plt.plot(sw,kro,label="Kro(Oil Relative Permeability)",color="green")


plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
#Step 12: Title
plt.figure(figsize=(12,8))

plt.plot(sw,krw,"*",label="Krw(Water Relative Permeability)",color='blue')
plt.plot(sw,kro,label="Kro(Oil Relative Permeability)",color="green")


plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()

plt.title("Relative Permeability vs Water Saturation")
```

```python
plt.style.available
```

```python
#Step 13: Stylish Plots 

plt.style.use("classic")

plt.figure(figsize=(12,8))

plt.plot(sw,krw,"*",label="Krw(Water Relative Permeability)",color='blue')
plt.plot(sw,kro,label="Kro(Oil Relative Permeability)",color="green")


plt.xlabel("Water Saturation,Sw(fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()

plt.title("Relative Permeability vs Water Saturation")
```
