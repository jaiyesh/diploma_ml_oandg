### Assignment: Write code, where porosity and perm thresholds will be given by user and you will return final dataframe

```python
df
```

```python
def poro_perm_filter(df):
    poro_thresh = float(input("Porosity Threshold: "))
    perm_thresh = float(input('Perm Threshold: '))
    
    df = df[(df["poro(%)"]>=poro_thresh)&(df["Perm(mD)"]>=perm_thresh)]
    
    return df
```

## Visualization

- Matplotlib

```python
import matplotlib.pyplot as plt 
```

```python
sw = np.linspace(0,1,101)
```

```python
sw
```

```python
Krw = sw**3
```

```python
Krw
```

```python
Kro = (1-sw)**2
```

```python
Kro
```

```python
# Step 1: Krw vs sw plot

plt.plot(sw,Krw)
```

```python
# Step 2: Kro vs sw plot
plt.plot(sw,Kro)
```

```python
# Step 3: Krw and Kro both on same plot
plt.plot(sw,Krw)
plt.plot(sw,Kro)
```

```python
# Step 4: Give X axis Label 

plt.plot(sw,Krw)
plt.plot(sw,Kro)


plt.xlabel("Water Saturation(Fraction)")
```

```python
# Step 5: Give Y axis Label 

plt.plot(sw,Krw)
plt.plot(sw,Kro)

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")
```

```python
# Step 6: Grid
plt.plot(sw,Krw)
plt.plot(sw,Kro)

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
```

```python
# Step 7: Adding Legends

plt.plot(sw,Krw, label = "Krw(water rel pem)")
plt.plot(sw,Kro, label = "Kro(oil rel pem)")

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 8: Changing the size of the plot
plt.figure(figsize=(12,8))
plt.plot(sw,Krw,label = "Krw (water Rel Perm)")
plt.plot(sw,Kro,label = "Kro (Oil Rel Perm)")

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 9: Changing the line color:
plt.figure(figsize=(12,8))
plt.plot(sw,Krw,label = "Krw (water Rel Perm)",color='blue')
plt.plot(sw,Kro,label = "Kro (Oil Rel Perm)",color ='brown')

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 10: Changing line thickness
plt.figure(figsize=(12,8))
plt.plot(sw,Krw,label = "Krw (water Rel Perm)",color='blue',lw=0.6)
plt.plot(sw,Kro,label = "Kro (Oil Rel Perm)",color ='brown',lw=4)

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 11: scatters
plt.figure(figsize=(12,8))
plt.plot(sw,Krw,label = "Krw (water Rel Perm)",color='blue',lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw,Kro,label = "Kro (Oil Rel Perm)",color ='brown',lw=4)
plt.scatter(sw,Kro)

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
```

```python
# Step 12: TITLE of the plot
plt.figure(figsize=(12,8))
plt.plot(sw,Krw,label = "Krw (water Rel Perm)",color='blue',lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw,Kro,label = "Kro (Oil Rel Perm)",color ='brown',lw=4)
plt.scatter(sw,Kro)

plt.xlabel("Water Saturation(Fraction)")
plt.ylabel("Relative Permeabilites(Fraction)")

plt.grid(True)
plt.legend()
plt.title('REL PERM PLOT')
```

```python
# Step 14: limit x and y axis

plt.figure(figsize=(12,8))
plt.plot(sw,Krw,label = "Krw (water Rel Perm)",color='blue',lw=0.6)
plt.scatter(sw,Krw)
plt.plot(sw,Kro,label = "Kro (Oil Rel Perm)",color ='brown',lw=4)
plt.scatter(sw,Kro)

plt.xlabel("Water Saturation(Fraction)")
plt.xlim(0,1)
plt.ylabel("Relative Permeabilites(Fraction)")
plt.ylim(0,1)

plt.grid(True)
plt.legend()
plt.title('REL PERM PLOT')
```
